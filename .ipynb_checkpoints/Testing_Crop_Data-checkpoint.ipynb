```python
img_path=r"test/acer_campestre/ny1081-02-1.jpg"
img=cv2.imread(img_path)

#visualizing img Image
if img is None:
    print("Error: Image not loaded correctly. Check the file path.")
else:
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Visualize the image
    plt.imshow(img_rgb)
    plt.title("Test Image")
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

### Testing Model

```python
image=tf.keras.preprocessing.image.load_img(img_path,target_size=(128,128))
input_arr=tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])
print(input_arr,input_arr.shape)
```

```python
prediciton=model.predict(input_arr)
print(prediciton,prediciton.shape)
```

```python
result_index=np.argmax(prediciton)
result_index
```

```python
class_name=["abies_concolor",
"abies_nordmanniana",
"acer_campestre","acer_ginnala"]
```

```python
#DisplayResult
model_prediction=class_name[result_index]
plt.imshow(img)
plt.title(f"Plant Species: {model_prediction}")
plt.xticks([])
plt.yticks([])
plt.show()
```
